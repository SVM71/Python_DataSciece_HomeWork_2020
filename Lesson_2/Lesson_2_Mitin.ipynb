```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')

%matplotlib inline

plt.style.use('fivethirtyeight')

cfd_data = pd.read_csv("~/py_DS/Lesson_2/310_23498_compressed_creditcard/creditcard.csv")
#cfd_data.info()

sC_vc = cfd_data['Class'].value_counts()
print(f"cfd_data['Class'].value_counts():\n{sC_vc}", '\n','='*20)

sC_vc.plot(kind="bar", logy=False)
plt.title("cfd_data['Class'].value_counts() - линейный")
plt.show()
sC_vc.plot(kind="bar", logy=True)
plt.title("cfd_data['Class'].value_counts() - логарифмический")
plt.show()
```

```python
# Вернет распределение значений последовательности в процентах
def convert_2_density( series_data, bins = 20 ):
    v1_4s = series_data.values
    v1_4s.sort()
    low_lim = np.floor(v1_4s[0])
    hi_lim = np.ceil(v1_4s[len(v1_4s)-1])
    delta = (hi_lim - low_lim)/bins
    #print(low_lim, hi_lim, delta)
    x = []
    dens_arr = []
    for i in range(bins):
        lb = low_lim + i*delta
        hb = lb + delta
        x.append(lb)
        dens_arr.append( round( 100.0 * sum( list( map( lambda  x: ( (x >= lb) and (x < hb ) ), v1_4s )))/len(v1_4s), 3 ) )
    res = pd.Series(dens_arr, index = x)
    return res

dv1 = convert_2_density(v1)
dv2 = convert_2_density(v2)
dv1.plot(kind='bar', logy = True)
plt.show()
dv2.plot(kind='bar', logy = True)
plt.show()

```
